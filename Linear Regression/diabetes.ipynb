```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
```

```python
data = load_diabetes()
```

```python
print(data)
```

```python
dataset = pd.DataFrame(data.data)
```

```python
dataset.head()
```

```python
dataset.columns = data.feature_names
```

```python
dataset
```

```python
X = dataset
Y = data.target
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,  test_size= 0.30, random_state= 42)
```

```python
X_train
```

```python
Y_train
```

```python
reg = LinearRegression()
reg.fit(X_train,Y_train)
```

```python
pred = reg.predict(X_train)
pred
```

```python
import seaborn as sns
sns.displot(pred - Y_train)
sns.displot(pred - Y_train, kind = 'kde')
```

```python
pred1 = reg.predict(X_test)
pred1
```

```python
sns.displot(pred1 - Y_test)
sns.displot(pred1 - Y_test, kind = 'kde')
```

```python
from sklearn.metrics import r2_score
score = r2_score(pred, Y_train)
score
```

```python
score1 = r2_score(pred1, Y_test)
score1
```
