```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
years_exp = np.array([1,3,5,6,4])
salary = np.array([31000, 33000, 34000, 37000, 38000])
```

```python
data = pd.DataFrame({'Experience in years':years_exp,'Salary': salary})
data
```

```python
mean_x = np.mean(years_exp)
mean_y = np.mean(salary)
```

```python
numer = 0
denom = 0
for i in range(len(years_exp)):
    numer += (years_exp[i] - mean_x) * (salary[i] - mean_y)
    denom += (years_exp[i] - mean_x) ** 2

m = numer/denom
b = mean_y - m * mean_x 
m = round(m,2)
b = round(b,2)
```

```python
m,b
```

```python
plt.scatter(years_exp, salary, color = 'blue', label = 'Data points')

reg_line = m * years_exp + b
plt.plot(years_exp,reg_line, color = 'Red', label = f'Line: y = {m}x + {b}')

plt.xlabel('Experience in years')
plt.ylabel('Salary')
plt.title('Years of Experience vs. Salary')
plt.legend()
plt.show()
```
